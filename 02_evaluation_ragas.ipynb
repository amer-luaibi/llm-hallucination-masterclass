# 📊 Module 2: Evaluating Hallucinations with Ragas
In this module, we use statistical and mathematical metrics to scientifically prove that the bot is hallucinating. We will measure **Faithfulness** and **Answer Relevance**.

```python
import os
from datasets import Dataset
from ragas import evaluate
from ragas.metrics import faithfulness, answer_relevance

# Set your OpenAI API Key
os.environ["OPENAI_API_KEY"] = "YOUR_API_KEY_HERE"
```

```python
# 1. Create the Evaluation Dataset
# We use the exact outputs generated by our broken chatbot in Module 1.
eval_data = {
    "question": [
        "What is the annual limit for dental check-ups?",
        "Does the premium plan cover full facial plastic surgery if I pay double the deductible?"
    ],
    "answer": [
        "The premium health insurance plan covers routine dental check-ups up to $500 annually.",
        "Yes, if you pay double the deductible, the premium plan may cover your facial plastic surgery depending on approval." # The Hallucinated Answer
    ],
    "contexts": [
        ["Our premium health insurance plan covers routine dental check-ups up to $500 annually. It does not cover cosmetic surgeries..."],
        ["Our premium health insurance plan covers routine dental check-ups up to $500 annually. It does not cover cosmetic surgeries..."]
    ]
}

dataset = Dataset.from_dict(eval_data)
```

```python
# 2. Run the Scientific Evaluation
print("Running scientific evaluation with Ragas...")

result = evaluate(
    dataset,
    metrics=[faithfulness, answer_relevance]
)
```

```python
# 3. Display Results
df = result.to_pandas()
print("\n--- Evaluation Results ---")
print(df[['question', 'faithfulness', 'answer_relevance']])
```
