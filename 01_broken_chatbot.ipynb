# 🛡️ Module 1: The Broken Chatbot (No Guardrails)
In this notebook, we build a standard RAG system. It works well for normal questions but fails terribly (hallucinates) when pushed outside its context.

```python
# 1. Install necessary libraries
# !pip install langchain openai chromadb tiktoken
```

```python
import os
from langchain.document_loaders import TextLoader
from langchain.indexes import VectorstoreIndexCreator
from langchain.chat_models import ChatOpenAI

# Set your OpenAI API Key
os.environ["OPENAI_API_KEY"] = "YOUR_API_KEY_HERE"
```

```python
# 2. Load the company knowledge base (RAG Setup)
print("Loading company policy data...")
loader = TextLoader("data/company_policy.txt")
index = VectorstoreIndexCreator().from_loaders([loader])
```

```python
# 3. Define the standard (unprotected) Chatbot
def ask_support_bot(question: str):
    print(f"\nUser Question: {question}")
    
    # The vulnerability: A basic prompt with no guardrails
    response = index.query(
        question, 
        llm=ChatOpenAI(temperature=0.7)
    )
    print(f"Bot Response: {response}")
    return response
```

```python
# 4. Testing the Bot (The Good and The Ugly)

# Test A: A normal question (Bot answers correctly based on context)
ask_support_bot("What is the annual limit for dental check-ups?")
```

```python
# Test B: The Hallucination Trigger (Out-of-context question)
# The bot will likely hallucinate a positive answer because there are no guardrails!
ask_support_bot("Does the premium plan cover full facial plastic surgery if I pay double the deductible?")
```
